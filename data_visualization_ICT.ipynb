```python
df10_x2017 = np.array(df10_y2017.columns.drop('Year'))
df10_y2017 = np.array(df10_y2017.drop('Year',axis=1)).reshape(len(df10_x2017))
df10_x2018 = np.array(df10_y2018.columns.drop('Year'))
df10_y2018 = np.array(df10_y2018.drop('Year',axis=1)).reshape(len(df10_x2018))
df10_x2019 = np.array(df10_y2019.columns.drop('Year'))
df10_y2019 = np.array(df10_y2019.drop('Year',axis=1)).reshape(len(df10_x2019))
df10_x2020 = np.array(df10_y2020.columns.drop('Year'))
df10_y2020 = np.array(df10_y2020.drop('Year',axis=1)).reshape(len(df10_x2020))
```
